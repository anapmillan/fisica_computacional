- Función <font color='blue'>range</font>: En Python, range es una función que genera una secuencia inmutable de números en un rango especificado. La sintaxis básica de la función range es range(start, stop, step). Donde:
    - start es el número inicial del rango (por defecto es 0).
    - stop es el número final del rango (no está incluido en la secuencia).
    - step es el tamaño del paso entre los números sucesivos (por defecto es 1).

La función range devuelve un objeto de tipo range, que puede ser utilizado en bucles for u otras situaciones donde se requiera una secuencia de números.


```python
for i in range(10, 3, -1):
    print(i)
```

### Otros métodos para listas en Python

1. **`list.append(x)`:** Agrega un ítem al final de la lista. **Equivalente a:** `a[len(a):] = [x]`
2. **`list.extend(iterable)`:** Extiende la lista agregándole todos los ítems del iterable. **Equivalente a:** `a[len(a):] = iterable`
3. **`list.insert(i, x)`:** Inserta un ítem en una posición dada. `a.insert(0, x)` inserta al principio. `a.insert(len(a), x)` equivale a `a.append(x)`.
4. **`list.remove(x)`:** Elimina el **primer** ítem de la lista cuyo valor sea `x`. **Lanza** `ValueError` si el ítem no existe.
5. **`list.pop([i])`:** Elimina y **retorna** el ítem en la posición `i`. Si `i` no se especifica, elimina y retorna el **último** elemento. **Lanza** `IndexError` si la lista está vacía o el índice está fuera de rango.
6. **`list.clear()`:** Elimina todos los elementos de la lista. **Equivalente a:** `del a[:]`
7. **`list.index(x[, start[, end]])`:** Retorna el índice basado en **cero** del **primer** elemento con valor `x`. **Lanza** `ValueError` si el elemento no existe. Los argumentos opcionales `start` y `end` limitan la búsqueda a un segmento específico. **Nota:** El índice retornado es relativo al inicio de la lista completa, **no** a `start`.
8. **`list.count(x)`:** Retorna el número de veces que `x` aparece en la lista.
9. **`list.sort(*, key=None, reverse=False)`:** Ordena los elementos de la lista **in situ**. Parámetros opcionales: `key`: Función personalizada para definir el criterio de ordenación. `reverse=True`: Ordena en orden descendente. Para más detalles, consulta `sorted()`.
10. **`list.reverse()`:** Invierte los elementos de la lista **in situ**.
11. **`list.copy()`:** Retorna una **copia superficial** de la lista. **Equivalente a:** `a[:]`

# Tuplas
En Python, una tupla es una colección ordenada e inmutable de elementos. Se definen utilizando paréntesis () y pueden contener cualquier tipo de datos, incluyendo otros contenedores como listas o incluso otras tuplas. A diferencia de las listas, las tuplas no pueden ser modificadas después de su creación.



Las tuplas son útiles cuando se necesita una colección de elementos que no cambiará durante la ejecución del programa, como coordenadas, claves de diccionario constantes o valores de configuración.





```python
# Definir una tupla
mi_tupla = (1, 2, 3, 'cuatro')

# Acceder a elementos de la tupla
print(mi_tupla[0])  # Output: 1
print(mi_tupla[3])  # Output: cuatro


# Crear una tupla de tuplas
tupla_anidada = ((1, 2), (3, 4), (5, 6))

# Acceder a elementos de la tupla anidada
print(tupla_anidada[0][1])  # Output: 2
print(tupla_anidada[2][0])  # Output: 5
```

# Gestión de memoria

En Python, la asignación de memoria y la declaración de variables funcionan de manera diferente a otros lenguajes de programación como C o Java. 

1. Asignación de Memoria:
    - En Python, cuando creamos un objeto (como un entero, una cadena de texto, una lista, etc.), Python reserva un espacio en la memoria para almacenar ese objeto.
    - Cada objeto en Python tiene un identificador único y una dirección de memoria asociada.
    - Cuando asignamos una variable a un objeto, en realidad estamos creando un enlace entre el nombre de la variable y la dirección de memoria donde se encuentra el objeto.
1. Variables como Punteros:
    - En Python, las variables son simplemente nombres que hacen referencia a objetos en la memoria. Es importante entender que las variables no contienen directamente los valores de los objetos, sino que actúan como punteros a la posición de memoria del objeto.
    - Cuando asignamos una variable a un objeto, estamos creando un puntero que apunta a la dirección de memoria donde se encuentra ese objeto.
1. Compartir Memoria entre Variables:
    - Para los objetos inmutables como escalares (x=4, y =2.5, f = True) y strings (c = 'casa') hay una asignación uno a uno entre variable y objeto (o espacio en la memoria). Cuando se asignan a nueva variable, se crea un nuevo objeto con un nuevo espacio en la memoria. 
    - Esto no es cierto para objetos inmutables como las listas (también se incluyen los diccionarios y los arrays de numpy). Si asignamos dos variables al mismo objeto, ambas variables apuntarán a la misma dirección de memoria. Esto significa que cualquier cambio realizado en uno de los objetos se reflejará en el otro, ya que ambos apuntan al mismo objeto en la memoria.
    - Este comportamiento afecta a las variables que apuntan a objetos mutables, como listas o diccionarios.

```python
# Ejemplo con listas
lista1 = [1, 2, 3]  # Crea una lista
lista2 = lista1     # Ambas variables apuntan a la misma lista
print(lista1, lista2)

```

```python
# Modificar lista2 afecta a lista1
lista2.append(4)
print(lista1, lista2)  # Output: [1, 2, 3, 4]
```

```python
# Ejemplo de asignación de variables inmutables: 
x = 10  # Asigna el valor 10 a la variable x
y = x   # La variable y apunta al mismo objeto que x

# Modificar y no afecta a x
y = 20
print(x)  # Output: 20
print(x,y)
```

```python
# Ejemplo con cadenas de caracteres: 
frase1 = 'original'
frase2 = frase1
del frase2
print(frase1)
# Las cadenas de caracteres son inmutables, no podemos cambiar frase2 una vez creada!
```

### Copia de Objetos:
- Para evitar que dos variables compartan la misma memoria, podemos realizar una copia del objeto utilizando métodos como `copy()` para listas, o funciones como `deepcopy()` para objetos más complejos.
- Esto crea un nuevo objeto en la memoria, y las variables apuntarán a direcciones de memoria diferentes.

```python
# Ejemplo con copia de lista
lista1 = [1, 2, 3]
lista2 = lista1.copy()  # Realiza una copia de la lista

# Modificar lista2 no afecta a lista1
lista2.append(4)
print(lista1)  # Output: [1, 2, 3]
print(lista2)
```

```python
# Lista anidada
lista_anidada = [[1, 2], [3, 4], [5, 6]]
print("Lista original:", lista_anidada)

# Hacer una copia superficial (shallow copy)
copia_superficial = lista_anidada.copy()

# Modificar la lista original
lista_anidada[0][0] = 100

# Imprimir las listas originales y la copia superficial
print("Lista cambiada:", lista_anidada)
print("Copia superficial:", copia_superficial)
```

```python
import copy

# Lista anidada
lista_anidada = [[1, 2], [3, 4], [5, 6]]
print("Lista original:", lista_anidada)

# Hacer una copia superficial (shallow copy)
copia_superficial = copy.copy(lista_anidada)

# Hacer una copia profunda (deep copy)
copia_profunda = copy.deepcopy(lista_anidada)

# Modificar la lista original
lista_anidada[0][0] = 100

# Imprimir las listas originales y las copias
print("Lista cambiada:", lista_anidada)
print("Copia superficial:", copia_superficial)
print("Copia profunda:", copia_profunda)
```
