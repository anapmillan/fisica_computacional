# Matplotlib

El paquete matplotlib es una extensa biblioteca de funciones para generar gráficos 2D y 3D. Es muy versátil. 

Matplotlib está construido sobre Numpy y trabaja de forma natural con arrays.

Dentro del paquete matplotlib destaca el módulo **pyplot**. Este módulo oculta muchas de las funcionalidades de bajo nivel de la biblioteca, facilitando el uso para crear gráficos sencillos. 



```python
import matplotlib.pyplot as plt
import numpy as np
```

## Gráficos básicos: función `plt.plot()`. 

- La función básica para representar es `plt.plot()`. Syntaxis:
    - `plt.plot(y)`
    - `plt.plot(x,y)`
    - `plt.plot(x,y,fmt)`
- Con la función `plt.show()` indicamos que se muestre por pantalla la figura. No siempre es necesaria, pero puede servir por ejemplo para refrescar una figura después de modificar los nombres de los ejes. 
- Podemos representar varias funciones superpuestas llamando repetidamente a la función `plt.plot()`.
- **Formato**: `fmt` es una cadena de caracteres con la que se puede especificar el estilo de línea y/o marcadores. Colores y formatos de línea: `params`: `plt.plot()` admite un tercer argumento opcional para indicar el estilo de representación con una cadena de caracteres, que indican el color ('b' para azul, 'g' para verde, 'k' para negro...) y el formato ('-' para línea continua, '--' para discontinua, ':' para línea de puntos, 'o', 's', '^' para puntos de distinta forma). En este formato se combinan las distintas opciones que se quiera indicar en una sola cadena de caracteres, por ejemplo: `'-k'` para una línea sólida negra, o `'or'` para círculos rojos. 
- Otros parámetros para controlar el estilo que se pueden introducir por referencia son:
    - `label`: Etiqueta para la línea que se utilizará en la leyenda. (Opcional)
    - `color`: Color de la línea y/o marcadores. Se puede especificar como un nombre de color (por ejemplo, 'blue', 'red', 'green'), un código hexadecimal (por ejemplo, '#FF5733'), o RGB (eg: [1,0,1]). (Opcional)
    - `linewidth`: Ancho de la línea. (Opcional)
    - `linestyle`: Estilo de línea, como '-' para sólido, '--' para discontinuo, ':' para punteado, etc. (Opcional)
    - `marker`: Tipo de marcador para los puntos, como 'o' para círculos, 's' para cuadrados, '^' para triángulos, etc. (Opcional)
    - `markersize`: Controla el tamaño de los marcadores.
    
## Edición de Gráficos 
Matplotlib proporciona diversas funciones y métodos para personalizar la apariencia de los gráficos, incluyendo la edición de etiquetas, títulos, rejilla (grid), escalas de los ejes, entre otros:

#### Etiquetas y Título
- `plt.xlabel(label)`: Establece la etiqueta del eje x.
- `plt.ylabel(label)`: Establece la etiqueta del eje y.
- `plt.title(label)`: Establece el título del gráfico.

#### Grid
- `plt.grid(bool)`: Habilita o deshabilita la rejilla en el gráfico. Se puede especificar si se desea mostrar la rejilla en los ejes x, y o ambos.


#### Creación de Subgráficos
- `fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))`: Crea una matriz de subgráficos con el número especificado de filas (nrows) y columnas (ncols). El tamaño de la figura se puede ajustar mediante el parámetro figsize.

#### Creación de Figuras
- `fig = plt.figure(figsize=(width, height))`: Crea una nueva figura para el gráfico con el tamaño especificado mediante el parámetro figsize.

#### Trabajar con ejes
Los objetos de eje `ax` (que devuelve por ejemplo plt.subplots) tienen muchos métodos útiles (a continuación se indican para el eje x, los métodos para el eje y son análogos). Muchos de ellos son análogos a las funciones de Pyplot, en el caso de una figura con varios subplots podemos indicar a qué plot nos referimos con estos métodos (en otros casos las funciones de representar admitirán un eje como argumento):
- `ax.set_xscale(scale)`: Establece la escala del eje x. Las escalas disponibles incluyen 'linear' (escala lineal), 'log' (escala logarítmica) y 'symlog' (escala logarítmica simétrica). 
- `ax.set_xlim(left=None, right=None)`: Establece los límites del eje x.
- `ax.set_xlabel(xlabel)`: Establece la etiqueta del eje x.
- `ax.set_title(label, fontdict=None, loc=None): Establece el título del gráfico. 
 

```python
# Representar con trazos
# Representa con trazos una secuencia de 10 números aleatorios, con abcisas de 0 a 9 
y = np.random.randint(0,10,10)
plt.plot(y)
```

```python
# Trazos y puntos
# Representa la función sen(x), con x de 0 a 4pi, mediante puntos unidos con trazos
# Cuántos puntos son necesarios para obtener una representación aceptable?
x = np.linspace(0,4*np.pi,9)
y = np.sin(x)
plt.plot(x,y,'o-')
```

```python
# Formatear estilo y color de línea
# Representa la secuencia de puntos siguiente mediante triángulos unidos por líneas de puntos

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y,'r^:')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas de puntos y Triángulos')
plt.show()
```

```python
# Utiliza ahora marcadores redondos de color rosa, de mayor tamaño, 
# y únelos con línea discontínua. 
# Incluye una leyenda, título, y labels en los ejes x e y.

plt.plot(x, y, color='pink', linestyle='--', marker='o', markersize=8, label='curva')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico con Líneas y Marcadores Personalizados')
plt.legend()
```

```python
# Representa las fucniones f1(x) = x, f2(x)=x**2, de x=1 a x=6. Incluye una leyenda. 
x = np.arange(1,6)
y1 = x**2
y2 = x

plt.plot(x, y1, label='f1')
plt.plot(x, y2, label='f2')
plt.legend()
```

## Representar nubes de puntos: 

La función `plt.scatter(x,y)` representa nubes de puntos sin unir por líneas. Simplifica algunas opciones de edición como el uso de mapas de colores para numbes de puntos. Sin ningún tipo de argumento opcional es equivalente a usar la función `plt.plot(x,y,'o')`.

Parámetros Principales que se pueden introducir por referencia:
- s: Tamaño de los marcadores. Puede ser un escalar que se aplicará a todos los marcadores o una secuencia del mismo tamaño que x e y que indica el tamaño de cada marcador individualmente. (Opcional)
- c: Color de los marcadores. Se puede especificar como un nombre de color (por ejemplo, 'blue', 'red', 'green') o un código hexadecimal (por ejemplo, '#FF5733'). (Opcional)
- marker: Tipo de marcador para representar cada punto, como 'o' para círculos, 's' para cuadrados, '^' para triángulos, etc. (Opcional)
- alpha: Transparencia de los marcadores, donde 0 es completamente transparente y 1 es completamente opaco. (Opcional)
- label: Etiqueta para los marcadores que se utilizará en la leyenda. (Opcional)
- cmap: mapa de colores (Opcional). 

```python
# Representa N=50 puntos distribuídos de forma aleatoria uniforme en 2D, 
# con coordenadas entre 0 y 1
# Añade título y labels
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión Básico')
plt.show()
```

```python
# Representa una nube de N=50 puntos con posiciones, tamaños y colores aleatorios.
# Incluye título, labels, y barra de escala de color

x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100
colors = np.random.rand(50)

# Podemos guardar el objeto resultante de plotear en "im":
im = plt.scatter(x, y, s=sizes, c=colors, marker='o', alpha=0.5, cmap = 'RdBu') 
# cmap controla el mapa de colores.
# por defecto: 'viridis'

# Añadimos una barra de colores con la funciòn plt.colorbar
# Como argumento toma un handle (puntero) a un objeto gráfico ya existente:
cbar = plt.colorbar(im)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión con Marcadores Personalizados y Colores')
plt.show()
```

```python
# Dos grupos de puntos
x1 = np.random.rand(50)
y1 = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.rand(50)

# Incluímos labels para la leyenda:
plt.scatter(x1, y1, label='Grupo 1')
plt.scatter(x2, y2, label='Grupo 2')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión con Etiquetas y Leyenda')
plt.legend()
plt.show()
```

## Representar histogramas

La función `plt.hist(x)` se utiliza para crear histogramas en Matplotlib, que muestran la distribución de datos agrupados en intervalos.

Parámetros Principales (por referencia):
- bins: Especifica cómo se agruparán los datos en intervalos. Puede ser un entero que indique el número de intervalos, una secuencia que defina los límites de los intervalos o una cadena que indique un método de cálculo automático de los intervalos. (Opcional)
- density: Si es True, el histograma mostrará una densidad de probabilidad normalizada en lugar de contar el número de muestras en cada bin. (Opcional, por defecto False)
- color: Color de las barras del histograma. Se puede especificar como un nombre de color (por ejemplo, 'blue', 'red', 'green') o un código hexadecimal (por ejemplo, '#FF5733'). (Opcional)
- alpha: Transparencia de las barras del histograma, donde 0 es completamente transparente y 1 es completamente opaco. (Opcional)
- label: Etiqueta para el histograma que se utilizará en la leyenda. (Opcional)

```python
# Histograma
data = np.random.randn(100000)

plt.hist(data, bins=30, color='blue')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma Básico')
plt.show()
```

```python
# Densidad de probabilidad

plt.hist(data, bins=30, density=True, color='green', alpha=0.7)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Histograma con Densidad de Probabilidad')
plt.show()
```

```python
# Combinación de varios histogramas
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  # Desplazar la segunda distribución

plt.hist(data1, bins=30, color='blue', alpha=0.5, label='Grupo 1')
plt.hist(data2, bins=30, color='red', alpha=0.5, label='Grupo 2')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma Múltiple con Leyenda')
plt.legend()
plt.show()
```

## Gráficos con barras de error 
La función `plt.errorbar(x,y,yerr,xerr,fmt)` se utiliza para crear gráficos de dispersión con barras de error. Esta función admite los parámetros por referencia usualues (color, linewidth, markersize, capsize). 


```python
# Gráfico con barras de error constantes
# Datos
x = np.linspace(0, 10, 40)
y = np.sin(x)
y_err = 0.1  # Magnitud del error en y

# Gráfico de dispersión con barras de error
plt.errorbar(x, y, yerr=y_err, fmt='o-', color='blue', markersize=5, capsize=5)

# Configuración adicional
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de dispersión con barras de error')
plt.grid(True)
```

```python
# Gráfico con barras de error distintas
# Datos
x = np.linspace(0, 10, 40)
y = np.sin(x)
y_err = 0.1*np.random.rand(40)  # Magnitud del error en y

# Gráfico de dispersión con barras de error
plt.errorbar(x, y, yerr=y_err, fmt='o', color='blue', markersize=5, capsize=5)

# Configuración adicional
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de dispersión con barras de error')
plt.grid(True)
```

## Gráficos de barras
La función `plt.bar(x,height, width=0.8, bottom=None)` se utiliza para crear gráficos de barras verticales. Es útil para comparar cantidades entre diferentes categorías. Otros parámetros que admite por referencia son:
- color: Color de las barras.
- edgecolor: Color del borde de las barras.
- alpha: Transparencia de las barras.
- align: Alineación de las barras respecto a sus valores x ('center', 'edge').
- label: Etiqueta para la leyenda.

```python
# Gráfico con barras
# Datos
x = np.arange(5)
altura = [10, 15, 7, 10, 12]

# Gráfico de barras
plt.bar(x, altura, color='green', alpha=0.5)

# Editamos el formato:
plt.xlabel('Categorías')
plt.ylabel('Altura')
plt.title('Gráfico de barras')
plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])  # Etiquetas del eje x
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Líneas de la cuadrícula en el eje y
plt.legend(['Altura'])
```

## Mapas de calor (heatmaps):
La función `plt.imshow(x)`.  Esta función toma como argumento una matriz de datos y opcionalmente varios parámetros para personalizar:

- cmap: Mapa de colores a utilizar para la visualización de la imagen.
- aspect: Proporción de aspecto de los ejes, que puede ser 'auto', 'equal' o un número que indica la relación entre las unidades de los ejes x e y.
- extent: Especifica los límites de los ejes de la imagen como (xmin, xmax, ymin, ymax).
- origin: Ubicación del origen de la imagen (esquina inferior izquierda o superior izquierda).
- interpolation: Método de interpolación utilizado para mejorar la calidad visual de la imagen.
- alpha: Valor de transparencia de la imagen.

```python
# Crear una matriz de datos aleatorios
data = np.random.rand(10, 10) +np.arange(10)


# Mostrar la imagen utilizando plt.imshow()
plt.imshow(data, cmap='gray', interpolation='nearest', vmin=0, vmax=10)
plt.colorbar()  # Mostrar la barra de color
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```
