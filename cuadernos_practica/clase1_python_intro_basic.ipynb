Además del indexado, también se admite el "slicing" (rebanado). Mientras que el indexado se utiliza para obtener caracteres individuales, el slicing te permite obtener una subcadena:

Los índices de slice tienen valores predeterminados útiles; un índice omitido por defecto es cero, y un índice omitido en segundo lugar es el tamaño de la cadena:

````python
word2[:2] # carácter desde el principio hasta la posición 2 (excluida)
````

````python
word2[4:7]
````

````python
word2[4:] # caracteres desde la posición 4 (incluida) hasta el final
````

````python
word2[-2:] # caracteres desde el penúltimo (incluido) hasta el final
````

Observa cómo el **inicio siempre está incluido y el final siempre está excluido**. Esto asegura que s[:i] + s[i:] siempre sea igual a s:

````python
print(word2[:2+1]) 
print(word2[2:])
````

Para índices no negativos, la longitud de un slice es la diferencia entre los índices, si ambos están dentro de los límites. Por ejemplo, la longitud de word[1:3] es 2.

**Intentar usar un índice que sea demasiado grande resultará en un error. Sin embargo, sí se pueden utilizar índices de slice fuera de rango:**

````python
word[6]
````

````python
word[4:42]
````

````python
word[42:]
````

## f-strings

Las *f-strings* en Python permiten insertar variables y expresiones dentro de cadenas de texto de forma sencilla y clara. Se escriben anteponiendo una `f` a la cadena y usando llaves `{}` para incluir el contenido dinámico, por ejemplo:


````python
# Opcion 1 :(((
area = 20/3
nombre1 = 'datos_area'+str(area)+'.txt'
print(nombre1)
````

````python
nombre1 = f'datos_area{area:.2f}.txt'
print(nombre1)
````

````python
frase = f'el área es {area} m^2'
print(frase)
print(f'el área es {area} m^2') 
````

### Funciones y Métodos para Strings

1. `print()` e `input()` son funciones incorporadas para interactuar con la entrada y salida estándar de un programa:
    1. `print(x)` muestra un mensaje por pantalla. Puede tomar uno o más argumentos separados por comas y los imprime en la consola, convirtiendo automáticamente los valores en strings si es necesario.
    2. `input(x)` se utiliza para obtener la entrada del usuario desde la consola. Muestra un mensaje opcional al usuario, espera a que ingrese datos y luego devuelve lo que el usuario escribe como una cadena de texto.    
2. En Python, un **método** es una función que pertenece a un objeto y puede realizar operaciones específicas en ese objeto.
3. En el caso de las cadenas de texto, los métodos realizan operaciones como dividirlas, eliminar espacios en blanco, contar caracteres, etc.
4. La sintaxis es:
    ```python
    cadena.metodo()
    ```
5. A veces los métodos requieren argumentos, otras no.

Éstos son algunos de los métodos más comunes para strings:

- `string.split(delimiter)`: Divide una cadena en subcadenas utilizando un separador específico. Si no se indica separador, utiliza el espacio como separador.
- `string.strip()`: Elimina los espacios en blanco u otros caracteres especificados del principio y final de una cadena.
- `string.count(x1)`: Cuenta el número de ocurrencias de una subcadena `x1` dentro de una cadena.
- `string.replace(x1, x2)`: Reemplaza todas las ocurrencias de una subcadena `x1` por otra especificada `x2`.
- `string.isupper()`: Devuelve `True` si todos los caracteres de la cadena son mayúsculas.
- `string.islower()`: Devuelve `True` si todos los caracteres de la cadena son minúsculas.
- `string.lower()`, `string.upper()`: Convierte los caracteres de la cadena en minúsculas o mayúsculas, respectivamente.


````python
# print e input
print("Hola mundo")
x = 10
print('El valor de x es', x)
````

````python
# Utiliza input para pedir el nombre del usuario. Después utilizar print para saludar al usuario por su nombre
a = input('Por favor ingrese un número')
print(a)
````

````python
# Utiliza string.split() para dividir la siguiente frase en palabras. Saca el resultado por pantalla:
frase = "Hola Mundo Python"
palabras = frase.split()
print(palabras)
````

````python
# Utiliza string.strip() para quitar los espacios al compienzo y final de la siguiente frase. Saca el resultado por pantalla:
cadena = "     Hola Mundo     "
cadena_limpia = cadena.strip()
print(cadena_limpia)
````

````python
# Utiliza string.count(caracter) para contar el número de veces que aparecen la palabra tigre y el string tr 
# en la siguiente frase. Saca el resultado por pantalla:
frase = "tres tristes tigres comen trigo en un trigal, ¿qué triste tigre come más?"

n_tigre = frase.count('tigre')
print(f'La palabra tigre aparece {n_tigre} veces')

mini_word = 'tr'
n_tr = frase.count(mini_word)
print(f'La cadena de caracteres {mini_word} aparece {n_tr} veces')
````

````python
# Utiliza string.replace(letra_original, letra_nueva) para reemplazar "Mundo" por "Python" en la siguiente frase. 
# Saca el resultado por pantalla:

frase = "Hola Mundo" 
frase.replace('Mundo', 'Python')
print(frase)

frase2 = frase.replace('Mundo', 'Python')
print(frase2)

frase = frase.replace('Mundo', 'Python')
print(frase)
````

````python
# Ejemplos con string.isupper(), string.islower(), string.upper(), string.lower()
cadena = "Hello Hello"
is_mayus = cadena.isupper()
is_minus = cadena.islower()
print("Mayúculas:", is_mayus)
print("Minúsculas:", is_minus)
print(cadena.upper().isupper())
print(cadena.lower())
````

````python
# Más ejemplos con string.isupper(), string.islower()
cadena2 = "HI"
cadena3 = "hola"
is_mayus2 = cadena2.isupper()
is_minus3 = cadena3.islower()
print("Mayúculas:", is_mayus2)
print("Minúsculas:", is_minus3)
````
